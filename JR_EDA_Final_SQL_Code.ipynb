```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("sqlite:///mta.db")
all_data = pd.read_sql('SELECT * FROM mta_data;', engine)
```

```python
#Since each row is one specific turnstile entries in a given time. Count each row and group by station to see "top 10 row entry":

top_10_station = pd.read_sql('select STATION, count(ENTRIES) as ENTRIES_COUNT from mta_data group by STATION order by ENTRIES_COUNT desc limit 10;', engine)
```

```python
top_10_station
```

```python
#See 34 ST-PENN STATION total row entries per day

daily_entries_34st_penn = pd.read_sql('select station, date, count(entries) as ROW_ENTRIES from mta_data where STATION="34 ST-PENN STA" group by DATE;',engine)
```

```python
daily_entries_34st_penn
```

```python
#Using subquery, compile row entries that are greater than 30,000 in all NYC stations

greater_30k=pd.read_sql('select DATE, ROW_ENTRIES from (select STATION, DATE, TIME, count(entries) as ROW_ENTRIES from mta_data group by DATE) where ROW_ENTRIES > 30000;',engine)
```

```python
greater_30k
```

```python
#Row entries of station that starts with G

g_station=pd.read_sql('select STATION, COUNT(ENTRIES) AS ROW_ENTRIES from mta_data where STATION like "G%" group by STATION;',engine)
```
